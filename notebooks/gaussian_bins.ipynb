```python
import torch

def create_gaussian_bins(mean, std, num_bins):
    # Create evenly spaced points in [0, 1]
    uniform_points = torch.linspace(0, 1, num_bins + 1)  # num_bins + 1 for edges
    # Define a normal distribution
    normal_dist = torch.distributions.Normal(mean, std)
    # Transform uniform points to the Gaussian scale
    bins = normal_dist.icdf(uniform_points)
    return bins

# Example usage
mean = 0.0
std = 1.0
num_bins = 10
bins = create_gaussian_bins(mean, std, 50)
```

```python
import matplotlib.pyplot as plt

plt.hist(bins[:-1], bins=bins, edgecolor='k')
plt.title('Bins with density centered around Gaussian mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
